```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('Merkez.csv')
ulkeler = veriSeti['Country']
nufus = veriSeti['Population']

plt.figure(figsize=(20, 7))  # Grafiğin boyutunu ayarlayın
plt.bar(ulkeler, nufus)
plt.xlabel('Ülkeler')
plt.ylabel('Nüfus')
plt.title('ülkelere Göre Nufüs Dağılımı')
plt.xticks(rotation=90)  # Ülke isimlerinin dikey yazılması
plt.show()
```
